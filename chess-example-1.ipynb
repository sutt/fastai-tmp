```python
import os, sys
import cv2
from pathlib import Path
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data_dir = Path('data/chess/')
list_imgs = []
for pic_fn in [e for e in os.listdir(data_dir) if 'WIN' in e]:
    list_imgs.append(cv2.imread(str(data_dir/pic_fn)))
```
